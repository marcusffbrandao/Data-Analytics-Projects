```python
# Imported modules:

from random import choice
import re
```

```python
# Creating the class Hangman:

class Hangman:
    def __init__(self, topic_name, topic_list = []):
        self.topic_name = topic_name
        self.topic_list = topic_list
        
    #=====================================================================================================            

    # Class variables used in the game:

    chances = 10
    current_word = ''
    secret_word = ''
    secret_word_list = []
    player_guess = ''
    player_letter = ''
           
        
    #=====================================================================================================            
 
    # Select secret word function:

    def select_word(self):  
        self.secret_word = choice(self.topic_list)
        self.secret_word_list = list(self.secret_word)
        for i in range(len(self.secret_word_list)):
            self.secret_word_list[i] = '_'    
            self.current_word = ' '.join(self.secret_word_list)
        print()
        print(f"Let's see if you really know about {self.topic_name}!")
        print()
        return self.current_word

    #=====================================================================================================            

    # Guess letter in the word function:
     
    def guess_letter(self):
    
        local_secret_word_list = list(self.secret_word)

        right_letter = 0

        print()
        print("Chose the letter you think the word contains.")
        print()
        self.player_letter = input()
        
        if len(re.findall('[a-zA-Z]', self.player_letter)) == 0:
            print()
            print('ATTENTION!!! Type one of the 26 letters of the English Alphabet!')
            print()
            return self.guess_letter()
        else:     
            self.player_letter = self.player_letter.lower()
            for i in range(len(self.secret_word_list)):
                if self.player_letter == local_secret_word_list[i]:
                    right_letter += 1
                    self.secret_word_list[i] = self.player_letter
                    self.current_word = ''.join(self.secret_word_list)
                if self.current_word == self.secret_word:
                    print()
                    print(f'You are right! The secret {self.topic_name} is {self.secret_word}!')
                    print()
                    self.chances = 0
                    self.hangman_draw("winner")
                    return self.chances
            if right_letter == 0:
                self.chances -= 1                  
                self.hangman_draw(self.chances)
            else:
                self.hangman_draw(self.chances)   
        return self.chances

    #=====================================================================================================            

    # Hangman drawing function:        
        
    def hangman_draw(self, arg):

        if arg == 10:
            print(f"                                      The {self.topic_name} has {len(self.secret_word)} letters:")      
            print()
            print(f"                                      {self.current_word}")
            print()
            
        elif arg == 9:

            print(f"                                      The {self.topic_name} has {len(self.secret_word)} letters:")      
            print(" _________________")
            print(f"|__H A N G M A N__|                   {self.current_word}")
            print()

        elif arg == 8:
            print()
            print("         |")
            print("         |")
            print("         |")
            print("         |")
            print("         |")
            print("         |")
            print("         |")
            print("         |")
            print("         |")
            print("         |")
            print("         |")      
            print(f"         |                            The {self.topic_name} has {len(self.secret_word)} letters:") 
            print(" ________|________")
            print(f"|__H A N G M A N__|                   {self.current_word}")
            print() 

        elif arg == 7:
            print("         ___________________")
            print("         |                 |")
            print("         |")
            print("         |")
            print("         |")
            print("         |")
            print("         |")
            print("         |")
            print("         |")
            print("         |")
            print("         |")
            print("         |")      
            print(f"         |                            The {self.topic_name} has {len(self.secret_word)} letters:") 
            print(" ________|________")
            print(f"|__H A N G M A N__|                   {self.current_word}")
            print() 

        elif arg == 6:
            print("         ___________________")
            print("         |                 |")
            print("         |               /////")
            print("         |              (|õ õ|)")
            print("         |               \_~_/ ")
            print("         |")
            print("         |")
            print("         |")
            print("         |")
            print("         |")
            print("         |")
            print("         |")      
            print(f"         |                            The {self.topic_name} has {len(self.secret_word)} letters:")
            print(" ________|________")
            print(f"|__H A N G M A N__|                   {self.current_word}")
            print() 

        elif arg == 5:
            print("         ___________________")
            print("         |                 |")
            print("         |               /////")
            print("         |              (|õ õ|)")
            print("         |               \_~_/")
            print("         |                 |")
            print("         |                 |")
            print("         |                 |")
            print("         |                 |")
            print("         |")
            print("         |")
            print("         |")      
            print(f"         |                            The {self.topic_name} has {len(self.secret_word)} letters:") 
            print(" ________|________")
            print(f"|__H A N G M A N__|                   {self.current_word}")
            print()   

        elif arg == 4:
            print("         ___________________")
            print("         |                 |")
            print("         |               /////")
            print("         |              (|õ õ|)")
            print("         |               \_~_/ ")
            print("         |                _|")
            print("         |               / |")
            print("         |              /  |")
            print("         |             {`  |")
            print("         |")
            print("         |")
            print("         |")      
            print(f"         |                            The {self.topic_name} has {len(self.secret_word)} letters:") 
            print(" ________|________")
            print(f"|__H A N G M A N__|                   {self.current_word}")
            print()      

        elif arg == 3:
            print("         ___________________")
            print("         |                 |")
            print("         |               /////")
            print("         |              (|õ õ|)")
            print("         |               \_~_/ ")
            print("         |                _|_")
            print("         |               / | \ ")
            print("         |              /  |  \ ")
            print("         |             {`  |  ´}")
            print("         |")
            print("         |")
            print("         |")      
            print(f"         |                            The {self.topic_name} has {len(self.secret_word)} letters:") 
            print(" ________|________")
            print(f"|__H A N G M A N__|                   {self.current_word}")
            print()  


        elif arg == 2:
            print("         ___________________")
            print("         |                 |")
            print("         |               /////")
            print("         |              (|õ õ|)")
            print("         |               \_~_/ ")
            print("         |                _|_")
            print("         |               / | \ ")
            print("         |              /  |  \ ")
            print("         |             {`  |  ´}")
            print("         |                /   ")
            print("         |               /     ")
            print("         |             _/")      
            print(f"         |                            The {self.topic_name} has {len(self.secret_word)} letters:") 
            print(" ________|________")
            print(f"|__H A N G M A N__|                   {self.current_word}")
            print()

        elif arg == 1:
            print("         ___________________")
            print("         |                 |")
            print("         |               /////")
            print("         |              (|õ õ|)")
            print("         |               \_~_/ ")
            print("         |                _|_")
            print("         |               / | \ ")
            print("         |              /  |  \ ")
            print("         |             {`  |  ´}")
            print("         |                / \ ")
            print("         |               /   \ ")
            print("         |             _/     \_")      
            print(f"         |                            The {self.topic_name} has {len(self.secret_word)} letters:")
            print(" ________|________")
            print(f"|__H A N G M A N__|                   {self.current_word}")
            print()

        elif arg == 0:
            print("         ___________________")
            print("         |                 |")
            print("         |               /////")
            print("         |              (|x x|)")
            print("         |               \_O_/ ")
            print("         |")                  
            print("         |             GAME OVER!")    
            print("         |")                         
            print("         |                _|_")
            print("         |               / | \ ")
            print("         |              /  |  \ ")
            print("         |             {`  |  ´}")
            print(f"         |                / \          The secret {self.topic_name} was:")
            print(" ________|________       /   \ ")
            print(f"|__H A N G M A N__|    _/     \_       {self.secret_word}")      
            print()

        elif arg == 'winner':
            print("         ___________________")
            print("         |                 |")
            print("         |                                        /////")
            print("         |                                    () (|õ õ|) ")
            print("         |                                     \  \_~_/ ")
            print("         |                                      \___|_")
            print("         |                                          | \ ")
            print("         |                                          |  \ ")
            print("         |                                          |  ´}")
            print("         |                                         / \ ")
            print("         |                                        /   \ ")
            print(f"         |                                      _/     \_")      
            print(" ________|________")
            print(f"|__H A N G M A N__|                             YOU WIN!!!")
            print()
        
    #=====================================================================================================            
            
    # Guess secret word function:

    def guess_word(self):

        print(f'''
    Do you want to guess the secret word? 

    Remember that if you guess it wrongly, you lose the game!

    Yes = type "1" and press "enter"
    No = type "2" and press "enter"
    ''')
        guess_choice = input()   

        if guess_choice == '1':    
            print(f'Type the name of the {self.topic_name}:')
            player_guess = input()

            if player_guess.lower() == self.secret_word: 
                print('You win the game!')
                self.chances = 0
                self.hangman_draw('winner')
                return self.chances
            else:
                print()
                print("Sorry, you guessed it wrongly! You lost the game!")
                print()
                self.chances = 0
                self.hangman_draw(0)
                return self.chances
        elif guess_choice == '2':
            return self.guess_letter()
        else:
            print('ATTENTION!!! Type 1 for "yes" or 2 for "no"!')
            return self.guess_word()
    
    #=====================================================================================================            
                
    # Game function:

    def run_game(self):
        #self.welcome_text()
        self.select_word()
        self.hangman_draw(self.chances)
        while self.chances > 0:
            self.guess_word()
            print(f'You have {self.chances} chances left!')
            print()
            print('=======================================================================================')

#=========================================================================================================
   
# Creating objects "Fruits" and "Animals", instantianting the class "Hangman":

Fruit = Hangman('fruit', ['apple', 'avocado', 'banana', 'blueberry', 'cherry', 'grappe', 'lemon', 'mango', 'orange', 'strawberry'])
Animal = Hangman('animal', ['dog', 'cat', 'chimpanzee', 'giraffe', 'kangaroo', 'lion', 'raccoon', 'squirrel', 'tiger', 'zebra'])

#=========================================================================================================
     
# Greeting function:
  
def welcome_text():
    print('''Welcome to the Hangman game!
    
    Do you know to play?

    Yes = type "1" and press "enter"
    No = type "2" and press "enter"
    ''')
        
    know_the_rules = input()   

    if know_the_rules == '1':
        pass
    elif know_the_rules == '2':
        print('''This game consists in guessing the secret word.
The player selects a topic (animals, fruits etc.) and the computer chooses a name of an element of the chosen topic.
The word to be guessed is represented by a line of dashes, symbolizing each letter of the word.
At the start of each round, the player will be given the opportunity to guess the secret word.
If the guess is wrong, the player will lose the game.
If the player does not want to guess the word, he or she can choose the letter that he or she thinks the secret word contains.
If the word contains the guessed letter, the corresponding line will be replaced by the letter.
However, if the secret word does not contain the guessed letter, a man on a gallows will begin to be drawn.
The player wins if he or she guesses or completes the word before 10 (ten) rounds.
Otherwise, at the end of the tenth round, the man will be hung and the game will be over.               
''')
        pass
    else:
        welcome_text()

#=========================================================================================================

# Select topic function:
  
def select_topic():
            
    print('''
    Choose the topic you would like to play:
    
    Fruit = type "1" and press "enter"
    Animal = type "2" and press "enter"
    ''')
    
    player_topic = input() 
    
    if player_topic == '1':
        return Fruit.run_game()
    elif player_topic == '2':
        return Animal.run_game()  
    else:
        print('ATTENTION!!! Type the number of the topic you want to play! Example: type 1 for "fruit".')
        return select_topic()

#=========================================================================================================

# Function that starts the game:

def play_game():
    welcome_text()
    select_topic()
        
#=========================================================================================================
```
